```python
df = pd.read_csv('jadwal.csv')
```

```python
df.drop(['Hari'], inplace=True, axis=1)
```

```python
df['tahun'] = df['Tanggal'].apply(lambda x: int(x.split('-')[0]))
df['bulan'] = df['Tanggal'].apply(lambda x: int(x.split('-')[1]))
df['hari'] = df['Tanggal'].apply(lambda x: int(x.split('-')[2]))
```

```python
df.drop(["Tanggal"], inplace=True, axis=1)
```

```python
creds = initialize_credentials()

# Load csv
# Panggil fungsi create_assignment dari Assignment.py
for index, row in df.iterrows():
    description = f"""
    Belajar tentang di Leetcode sebanyak 10 soal
    {row['DSA Topic']} \n\n\n 
    Belajar tentang ML [coba untuk menerapkannya dari scratch tanpa library sklearn]: 
    {row['ML Topic']} \n\n 
    Latihan dengan voice assistant bisa chat GPT atau Google Gemini tentang:
    {row['System/Behavioral']}
    """
    create_assignment(
        creds,
        title=row['Motivation'],
        description=description,
        scheduledTime={
            'Tahun': row['tahun'],
            'Bulan': row['bulan'],
            'Hari': row['hari'],
            'Jam': 3,
            'Menit': 55
        },
        due_date={ 'year': row['tahun'], 'month': row['bulan'], 'day': row['hari'] },
        due_time={ 'hours': 16, 'minutes': 59 },
        # state='PUBLISHED' # DRAFT, PUBLISHED, DELETED
    )
    print(f"Created assignment for {row['Motivation']}")
    break
print("All assignments created.")
```
